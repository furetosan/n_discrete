```python
# Parameter values
# Masses:
# m1 = 2.0
m = 2.0/n
# Spring constants
# k1 = 5.0
k = 10.0*n
# Natural lengths
# L1 = 1.0
L = 4.0/n
# Friction coefficients
# b1 = 0.6
b = 91.5/n

# Initial conditions
# x1 and x2 are the initial displacements; y1 and y2 are the initial velocities
x1 = 1.0*L
y1 = 0.0
x2 = 2.0*L
y2 = 0.0
x3 = 3.0*L
y3 = 0.0
x4 = 4.0*L
y4 = 0.0
x5 = 5.0*L
y5 = 0.0
x6 = 6.0*L
y6 = 0.0
x7 = 7.0*L
y7 = 0.0
x8 = 8.0*L
y8 = 0.0
x9 = 9.0*L
y9 = 0.0
x10 = 10.0*L
y10 = 0.0
x11 = 11.0*L
y11 = 0.0
x12 = 12.0*L
y12 = 0.0
x13 = 13.0*L
y13 = 0.0
x14 = 14.0*L
y14 = 0.0
x15 = 15.0*L
y15 = 0.0
x16 = 16.0*L
y16 = 0.0

# ODE solver parameters
abserr = 1.0e-8
relerr = 1.0e-6
stoptime = 25.0
numpoints = 1000

# Create the time samples for the output of the ODE solver.
# I use a large number of points, only because I want to make
# a plot of the solution that looks nice.
t = [stoptime * float(i) / (numpoints - 1) for i in range(numpoints)]
```

#### Degrau de For√ßa de Entrada

```python
# degrau de forca
def degrau_forca(t):
    if t <= 5.0:
        forca = forca_inicial
    else:
        forca = forca_final
    return forca
```

#### Campo Vetorial

```python
def degrau_forca_vectorfield(w, t, p):
    """
    Defines the differential equations for the coupled spring-mass system.

    Arguments:
        w :  vector of the state variables:
                  w = [x1,y1,x2,y2,x3,y3]
        t :  time
        p :  vector of the parameters:
                  p = [m1,m2,m3,k1,k2,k3,L1,L2,L3,b1,b2,b3]
    """
    x1, y1, x2, y2, x3, y3, x4, y4, x5, y5, x6, y6, x7, y7, x8, y8, x9, y9, \
      x10, y10, x11, y11, x12, y12, x13, y13, x14, y14, x15, y15, x16, y16 = w
    m, k, L, b = p

    # Create f = (x1',y1',x2',y2' ...):
    f = [
            y1,
            (-b * y1 + b * (y2 - y1) - k * (x1 - L) + k * (x2 - x1 - L) ) / m,
            y2,
            (-b * (y2 - y1) + b * (y3 - y2) - k * (x2 - x1 - L) + k * (x3 - x2 - L) ) / m,
            y3,
         
            (-b * (y11 - y10) + b * (y12 - y11) - k * (x11 - x10 - L) + k * (x12 - x11 - L) ) / m,
            y12,
         
            (-b * (y14 - y13) + b * (y15 - y14) - k * (x14 - x13 - L) + k * (x15 - x14 - L) ) / m,
            y15,
            (-b * (y15 - y14) + b * (y16 - y15) - k * (x15 - x14 - L) + k * (x16 - x15 - L) ) / m,
            y16,
            (-b * (y16 - y15)  - k * (x16 - x15 - L) + degrau_forca(t)) / m,
        ]
    return f
```

```python
def degrau_forca_vectorfield(w, t, p):
    """
    Defines the differential equations for the coupled spring-mass system.

    Arguments:
        w :  vector of the state variables:
                  w = [x1,y1,x2,y2,x3,y3]
        t :  time
        p :  vector of the parameters:
                  p = [m1,m2,m3,k1,k2,k3,L1,L2,L3,b1,b2,b3]
    """
    x1, y1, x2, y2, x3, y3, x4, y4, x5, y5, x6, y6, x7, y7, x8, y8, x9, y9, \
      x10, y10, x11, y11, x12, y12, x13, y13, x14, y14, x15, y15, x16, y16 = w
    m, k, L, b = p

    # Create f = (x1',y1',x2',y2' ...):
    f = [
            y1,
            (-b * y1 + b * (y2 - y1) - k * (x1 - L) + k * (x2 - x1 - L) ) / m,
            y2,
            (-b * (y2 - y1) + b * (y3 - y2) - k * (x2 - x1 - L) + k * (x3 - x2 - L) ) / m,
            y3,
            (-b * (y3 - y2) + b * (y4 - y3) - k * (x3 - x2 - L) + k * (x4 - x3 - L) ) / m,
            y4,
            (-b * (y4 - y3) + b * (y5 - y4) - k * (x4 - x3 - L) + k * (x5 - x4 - L) ) / m,
            y5,
            (-b * (y5 - y4) + b * (y6 - y5) - k * (x5 - x4 - L) + k * (x6 - x5 - L) ) / m,
            y6,
            (-b * (y6 - y5) + b * (y7 - y6) - k * (x6 - x5 - L) + k * (x7 - x6 - L) ) / m,
            y7,
            (-b * (y7 - y6) + b * (y8 - y7) - k * (x7 - x6 - L) + k * (x8 - x7 - L) ) / m,
            y8,
            (-b * (y8 - y7) + b * (y9 - y8) - k * (x8 - x7 - L) + k * (x9 - x8 - L) ) / m,
            y9,
            (-b * (y9 - y8) + b * (y10 - y9) - k * (x9 - x8 - L) + k * (x10 - x9 - L) ) / m,
            y10,
            (-b * (y10 - y9) + b * (y11 - y10) - k * (x10 - x9 - L) + k * (x11 - x10 - L) ) / m,
            y11,
            (-b * (y11 - y10) + b * (y12 - y11) - k * (x11 - x10 - L) + k * (x12 - x11 - L) ) / m,
            y12,
            (-b * (y12 - y11) + b * (y13 - y12) - k * (x12 - x11 - L) + k * (x13 - x12 - L) ) / m,
            y13,
            (-b * (y13 - y12) + b * (y14 - y13) - k * (x13 - x12 - L) + k * (x14 - x13 - L) ) / m,
            y14,
            (-b * (y14 - y13) + b * (y15 - y14) - k * (x14 - x13 - L) + k * (x15 - x14 - L) ) / m,
            y15,
            (-b * (y15 - y14) + b * (y16 - y15) - k * (x15 - x14 - L) + k * (x16 - x15 - L) ) / m,
            y16,
            (-b * (y16 - y15)  - k * (x16 - x15 - L) + degrau_forca(t)) / m,
        ]
    return f
```


## Entrada de Rampa em For√ßa em $x_3$

### Entradas

#### Rampa de For√ßa de Entrada

```python
# rampa de forca
def rampa_forca(t):
    if t < tempo_inicial:
        forc = forca_inicial
    elif t >= tempo_inicial and t <= tempo_final:
        # forc = 5.0-1.0*t
        forc = forca_inicial+(forca_final-forca_inicial)/(10.0-5.0)*(t-5.0)
    else:
        forc = forca_final
    return forc
```

#### Campo Vetorial

```python
def rampa_forca_vectorfield(w, t, p):
    """
    Defines the differential equations for the coupled spring-mass system.

    Arguments:
        w :  vector of the state variables:
                  w = [x1,y1,x2,y2,x3,y3]
        t :  time
        p :  vector of the parameters:
                  p = [m1,m2,m3,k1,k2,k3,L1,L2,L3,b1,b2,b3]
    """
    x1, y1, x2, y2, x3, y3, x4, y4, x5, y5, x6, y6, x7, y7, x8, y8, x9, y9, \
      x10, y10, x11, y11, x12, y12, x13, y13, x14, y14, x15, y15, x16, y16 = w
    m, k, L, b = p

    # Create f = (x1',y1',x2',y2' ...):
    f = [
            y1,
            (-b * y1 + b * (y2 - y1) - k * (x1 - L) + k * (x2 - x1 - L) ) / m,
            y2,
            (-b * (y2 - y1) + b * (y3 - y2) - k * (x2 - x1 - L) + k * (x3 - x2 - L) ) / m,
            y3,
            (-b * (y3 - y2) + b * (y4 - y3) - k * (x3 - x2 - L) + k * (x4 - x3 - L) ) / m,
            y4,
            (-b * (y4 - y3) + b * (y5 - y4) - k * (x4 - x3 - L) + k * (x5 - x4 - L) ) / m,
            y5,
            (-b * (y5 - y4) + b * (y6 - y5) - k * (x5 - x4 - L) + k * (x6 - x5 - L) ) / m,
            y6,
            (-b * (y6 - y5) + b * (y7 - y6) - k * (x6 - x5 - L) + k * (x7 - x6 - L) ) / m,
            y7,
            (-b * (y7 - y6) + b * (y8 - y7) - k * (x7 - x6 - L) + k * (x8 - x7 - L) ) / m,
            y8,
            (-b * (y8 - y7) + b * (y9 - y8) - k * (x8 - x7 - L) + k * (x9 - x8 - L) ) / m,
            y9,
            (-b * (y9 - y8) + b * (y10 - y9) - k * (x9 - x8 - L) + k * (x10 - x9 - L) ) / m,
            y10,
            (-b * (y10 - y9) + b * (y11 - y10) - k * (x10 - x9 - L) + k * (x11 - x10 - L) ) / m,
            y11,
            (-b * (y11 - y10) + b * (y12 - y11) - k * (x11 - x10 - L) + k * (x12 - x11 - L) ) / m,
            y12,
            (-b * (y12 - y11) + b * (y13 - y12) - k * (x12 - x11 - L) + k * (x13 - x12 - L) ) / m,
            y13,
            (-b * (y13 - y12) + b * (y14 - y13) - k * (x13 - x12 - L) + k * (x14 - x13 - L) ) / m,
            y14,
            (-b * (y14 - y13) + b * (y15 - y14) - k * (x14 - x13 - L) + k * (x15 - x14 - L) ) / m,
            y15,
            (-b * (y15 - y14) + b * (y16 - y15) - k * (x15 - x14 - L) + k * (x16 - x15 - L) ) / m,
            y16,
            (-b * (y16 - y15) - k * (x16 - x15 - L) + rampa_forca(t)) / m,
        ]
    return f
```

```python
def rampa_desl_vectorfield(w, t, p):
    """
    Defines the differential equations for the coupled spring-mass system.

    Arguments:
        w :  vector of the state variables:
                  w = [x1,y1,x2,y2,x3,y3]
        t :  time
        p :  vector of the parameters:
                  p = [m1,m2,m3,k1,k2,k3,L1,L2,L3,b1,b2,b3]
    """
    x1, y1, x2, y2, x3, y3, x4, y4, x5, y5, x6, y6, x7, y7, x8, y8, x9, y9, \
      x10, y10, x11, y11, x12, y12, x13, y13, x14, y14, x15, y15, = w # x16, y16
    m, k, L, b = p


    # Create f = (x1',y1',x2',y2' ...):
    f = [
            y1,
            (-b * y1 + b * (y2 - y1) - k * (x1 - L) + k * (x2 - x1 - L) ) / m,
            y2,
            (-b * (y2 - y1) + b * (y3 - y2) - k * (x2 - x1 - L) + k * (x3 - x2 - L) ) / m,
            y3,
            (-b * (y3 - y2) + b * (y4 - y3) - k * (x3 - x2 - L) + k * (x4 - x3 - L) ) / m,
            y4,
            (-b * (y4 - y3) + b * (y5 - y4) - k * (x4 - x3 - L) + k * (x5 - x4 - L) ) / m,
            y5,
            (-b * (y5 - y4) + b * (y6 - y5) - k * (x5 - x4 - L) + k * (x6 - x5 - L) ) / m,
            y6,
            (-b * (y6 - y5) + b * (y7 - y6) - k * (x6 - x5 - L) + k * (x7 - x6 - L) ) / m,
            y7,
            (-b * (y7 - y6) + b * (y8 - y7) - k * (x7 - x6 - L) + k * (x8 - x7 - L) ) / m,
            y8,
            (-b * (y8 - y7) + b * (y9 - y8) - k * (x8 - x7 - L) + k * (x9 - x8 - L) ) / m,
            y9,
            (-b * (y9 - y8) + b * (y10 - y9) - k * (x9 - x8 - L) + k * (x10 - x9 - L) ) / m,
            y10,
            (-b * (y10 - y9) + b * (y11 - y10) - k * (x10 - x9 - L) + k * (x11 - x10 - L) ) / m,
            y11,
            (-b * (y11 - y10) + b * (y12 - y11) - k * (x11 - x10 - L) + k * (x12 - x11 - L) ) / m,
            y12,
            (-b * (y12 - y11) + b * (y13 - y12) - k * (x12 - x11 - L) + k * (x13 - x12 - L) ) / m,
            y13,
            (-b * (y13 - y12) + b * (y14 - y13) - k * (x13 - x12 - L) + k * (x14 - x13 - L) ) / m,
            y14,
            (-b * (y14 - y13) + b * (y15 - y14) - k * (x14 - x13 - L) + k * (x15 - x14 - L) ) / m,
            y15,
            (-b * (y15 - y14) + b * (deriv_desl(t) - y15) - k * (x15 - x14 - L) + k * (rampa_disp(t) - x15 - L) ) / m,
            # (-b * y15 - k * (x15 - x14 - L) + k * (rampa_disp(t) - x15 - L) ) / m,
            # y16,
            # (-b * y16 - k * (rampa_disp(t) - x15 - L) ) / m,
        ]
    return f
```
